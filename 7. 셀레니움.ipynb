```python
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
driver.get("https://comic.naver.com/webtoon/detail?titleId=823195&no=1&week=finish")
time.sleep(2)
print("*** 베스트 댓글 ***")
# /html/body/div[1]/div[5]/div/div/div[4]/div[1]/div[3]/div/section/ul/li[1]  # full xpath 박스전체 (큰 틀)
# /html/body/div[1]/div[5]/div/div/div[4]/div[1]/div[3]/div/section/ul/li[2]
# /html/body/div[1]/div[5]/div/div/div[4]/div[1]/div[3]/div/section/ul/li[3]
xpath = '/html/body/div[1]/div[5]/div/div/div[4]/div[1]/div[3]/div/section/ul/li'
best_comment_elements = driver.find_elements(By.XPATH, xpath)
best_comment_elements

for li in best_comment_elements:
    try:
        # //*[@id="wcc_root"]/section/ul/li[1]/div[1]/div[2]/div/p  # 베스트 댓글 그 자체의 경로  # xpath 로 카피
        # //*[@id="wcc_root"]/section/ul/li[2]/div[1]/div[2]/div/p
        comment_p = li.find_element(By.XPATH, "div[1]/Div[2]/div/p")
        comment_text = comment_p.text.strip()
        print(comment_text)
        print("-" * 30)
    except Exception as e:
        print(e)
```

```python
from urllib.request import Request, urlopen
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome()
url = "https://pixabay.com/images/search/%ea%b0%95%ec%95%84%ec%a7%80/"
driver.get(url)

image_xpath = '/html/body/div[1]/div[1]/div/div[2]/div[3]/div/div/div[3]/div[1]/div/div/a/img'

wait = WebDriverWait(driver, 10)  # 최대 10초까지 기다림
image_element = wait.until(EC.presence_of_element_located((By.XPATH, image_xpath)))

image_url = image_element.get_attribute('src')

image_byte = Request(image_url, headers={'User-Agent' : 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_7)'})  # 사람인것 처럼 헤더 값을 보내야 한다.
image_byte

with open('dog2.jpg', 'wb') as f:  # wb 는 w 는 write / b 는 Binarry (글씨가 아닌것)
    f.write(urlopen(image_byte).read())
print('다운로드 완료!!')

driver.quit()
```

# **4. 야놀자 리뷰 크롤링**


```python
def crawl_yanolja_reviews(name, url, max_scroll=100):
    review_list = []
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(3)

    last_height = driver.execute_script("return document.body.scrollHeight")  # 파이썬에서 자바스크리브 언어를 사용하는 것
    # print(last_height)
    same_count = 0
    scroll_times = 0

    while True:
        scroll_times += 1
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")  # 파이썬에서 자바스크리브 언어를 사용하는 것
        time.sleep(2)
        new_height = driver.execute_script("return document.body.scrollHeight")

        if new_height == last_height:
            same_count += 1
        else : 
            same_count = 0
            last_height = new_height

        if same_count >= 3:
            print('더 이상 새로운 리뷰가 없습니다.')
            break

        if scroll_times >= max_scroll:
            print('최대 스크롤 횟수 도달. 종료합니다.')
            break

    # /html/body/div[1]/section/div/div[5]/div[1]
    # /html/body/div[1]/section/div/div[5]/div[2]
    # /html/body/div[1]/section/div/div[5]/div[1]/div[2]/div/div[3]/div[1]/p
    # /html/body/div[1]/section/div/div[5]/div[2]/div[2]/div/div[3]/div[1]/p

    containers = driver.find_elements(By.XPATH, "/html/body/div[1]/section/div/div[5]/div")
    # print('총 리뷰 개수 : ', len(containers))


    for c in containers:
        try :
            review_text = c.text.strip()
        except :
            review_text = ""
        # print(review_text)

        try : 
            # /html/body/div[1]/section/div/div[5]/div[1]/div[2]/div/div[1]/div[1]/div[2]/p
            # /html/body/div[1]/section/div/div[5]/div[2]/div[2]/div/div[1]/div[1]/div[2]/p
            date_text = c.find_element(By.XPATH,".//div/div/div/div/p").text
        except:
            date_text = ""
        print(date_text)

        review_list.append({
            "review" : review_text,
            "date" : date_text
        })

    driver.quit()
    return review_list
```
